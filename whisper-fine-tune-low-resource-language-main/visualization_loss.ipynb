```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def load_csv(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"{file_path} not found.")
    return pd.read_csv(file_path)[['Step', 'Value']]

def auto_range(series, buffer_ratio=0.05):
    """Calculate min and max range with buffer."""
    min_val = series.min()
    max_val = series.max()
    buffer = (max_val - min_val) * buffer_ratio
    return min_val - buffer, max_val + buffer

def plot_training_vs_evaluation(train_df, eval_df):
    train_df = train_df.copy()
    eval_df = eval_df.copy()
    train_df['Type'] = 'Training'
    eval_df['Type'] = 'Evaluation'
    merged_df = pd.concat([train_df, eval_df])

    # Calculate dynamic axis limits
    x_min, x_max = auto_range(merged_df['Step'])
    y_min, y_max = auto_range(merged_df['Value'])

    plt.figure(figsize=(10, 6))
    for label, group_df in merged_df.groupby('Type'):
        plt.plot(group_df['Step'], group_df['Value'], marker='o', linestyle='-', label=label)

    plt.xlabel("Batches Processed")
    plt.ylabel("Loss")
    plt.title("LoRA : Training vs Evaluation Loss")
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    training_csv = "loss_curves/lora_training.csv"
    evaluation_csv = "loss_curves/lora_evaluation.csv"

    train_df = load_csv(training_csv)
    eval_df = load_csv(evaluation_csv)

    plot_training_vs_evaluation(train_df, eval_df)
```
